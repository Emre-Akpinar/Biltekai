Cleaning last names by strip

There is another way to strip unwanted characters:

```python
df = pd.read_excel(r"C:\Users\akpin\PycharmProjects\Pandas\Customer Call List.xlsx")
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df = df.drop_duplicates()
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/._")
df
```

Cleaning/Standardizing phone numbers:

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

we need to take numbers as strings

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)[:3]+"-"+str(x)[3:6]+"-"+str(x)[6:])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--","",regex=True)
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--","",regex=True)
df
```

Address

```python
df[["Street_Address","State","Zip"]] = df["Address"].str.split(",",expand=True)
df
```

Standardizing column values on paying customer and do not contact

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y") 
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N") 
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y") 
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N") 
df
```

```python
df = df.replace("N/a","")
df = df.fillna("")
df
```

Filtering down Rows:

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x,inplace=True)

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace("","N")
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == "":
        df.drop(x,inplace=True)

df
#Another way to drop null values 
#df = df.dropna(subset="Phone_Number",inplace=True)
```
