## Changing the Index:

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\akpin\PycharmProjects\Pandas\world_population.csv")
df
```

#### There is many ways to to change the index:

### Using set_index method:

```python
df.set_index("Country",inplace=True)
df
```

### Reseting the data frame index:

```python
df.reset_index(inplace=True) #inplace=True means that we don't have to assign it to a new variable. It assigns it back to df.
df
```

### While read in:

```python
df = pd.read_csv(r"C:\Users\akpin\PycharmProjects\Pandas\world_population.csv",index_col="Country")
df
```

#### Searching through the index: 

```python
df.loc["Albania"]
```

```python
df.iloc[1]
```

# Multi indexing:

```python
df.reset_index(inplace=True)
df
```

```python
df.set_index(["Continent","Country"],inplace=True)
df
```

#### To make it look better:

```python
df.sort_index(inplace=True)
df
```
