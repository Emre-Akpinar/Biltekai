```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\akpin\PycharmProjects\Pandas\Ice Cream Ratings.csv")
df.set_index("Date", inplace=True)
df
```

## Plot function:

##### By default its a line graph

```python
df.plot()
```

### Separating the different (column)lines:

```python
df.plot(subplots=True)
```

### Setting title and axis names:

```python
df.plot(title = "Ice Cream Ratings")
```

```python
df.plot(title="Ice Cream Ratings",xlabel = "Daily Ratings",ylabel = "Scores")
```

## Different vizualization types:

### Bar:

```python
df.plot(kind = "bar")
```

#### Stacking the bar plots:

```python
df.plot(kind = "bar",stacked = True)
```

#### Specifying the column we want:

```python
df["Flavor Rating"].plot(kind = "bar")
```

### Horizontal bar:

```python
df.plot.barh()
```

#### Stacking horizontal bar:

```python
df.plot.barh(stacked = True)
```

### Scatter function

##### scatter() requires 2 positional argument(x and y).

```python
df.plot.scatter(x = "Texture Rating",y = "Overall Rating")
```

##### 's' for scale:

```python
df.plot.scatter(x = "Texture Rating",y = "Overall Rating",s = 100)
```

##### 'c' for color:

```python
df.plot.scatter(x = "Texture Rating",y = "Overall Rating",c = "green")
```

### hist()

```python
df.plot.hist(bins = 100) # bins is for distrubituon
```

### boxplot()

```python
df.boxplot()
```

### area()

```python
df.plot.area()
```

#### Adjust size:

```python
df.plot.area(figsize = (10,5))
```

### pie()

##### pie requires either y column or 'subplots=True'

```python
df.plot.pie(y = "Flavor Rating",figsize=(10,6))
```
